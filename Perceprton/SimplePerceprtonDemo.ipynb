## Simple Preceprton Trick

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
# Generate the dataset

import numpy as np
X, y = make_classification(n_samples=1000, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
X[:5]
```

```python
y[:5]
```

```python
df = pd.DataFrame(np.column_stack([X[:,0], X[:,0], y]), columns=['x1', 'x2', 'y'])

print("Shape: ",df.shape)

df.head()
```

```python
sns.scatterplot(data = df,x = 'x1',y = "x2",hue='y')
```

```python
# Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
print("Coeff:", p.coef_)
print("Intercept:", p.intercept_)
```

```python
#plot a decision boundary
plot_decision_regions(X, y, clf=p, legend=2)
```
